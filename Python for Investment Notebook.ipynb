# Create Short and Long EMA

###### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;EMA12 and EMA26: [https://www.financereference.com/ema12-vs-ema26/](https://www.financereference.com/ema12-vs-ema26/)

```python
data = data.assign(ema_short=data['Close'].ewm(span=12, adjust=False, min_periods=12).mean(),
                           ema_long=data['Close'].ewm(span=26, adjust=False, min_periods=26).mean())

data
```

```python
# Create EMA lines
ema_short_line = mpf.make_addplot(data['ema_short'], color='orange')
ema_long_line = mpf.make_addplot(data['ema_long'], color='blue')
```

```python
# Join two plots
ema_list = [ema_short_line, ema_long_line]
```

```python
# Plot graph
mpf.plot(data, style = 'yahoo', addplot = ema_list, type = 'candle')
```

<br>

# Analyzing when to buy or sell

###### - Create Trend Column
###### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; If the EMA12 line is greater than the EMA26 line, then the trends will show TRUE

```python
data['trend'] = data['ema_short'] > data['ema_long']
data
```

###### - Closing Price

```python
# Show normal closing price
data['Close']
```

```python
# Shift the closing price by 1 interval (closing price of the previous day)
data['Close'].shift(1)
```

###### - Create trend_shift Column

```python
data['trend_shift'] = data['trend'].shift(1)
data
```

###### - Analyze the Trend Shift
###### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 'buy' when the trend changes from downward to upward
###### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 'sell' when the trend changes from upward to downward

```python
data.loc[(data.trend == True) & (data.trend_shift == False), 'action'] = 'buy'
data.loc[(data.trend == False) & (data.trend_shift == True), 'action'] = 'sell'
data
```

```python
# Filter only 'buy'
data[data.action == 'buy']
```

```python
# Filter only 'sell'
data[data.action == 'sell']
```

```python
# Define mark point
#Create a gap while buying or selling by setting the 'marker_position' to 95% and 105% of the lowest and highest price
data.loc[data['action'] == 'buy', 'marker_position'] = data['Low']*0.95
data.loc[data['action'] == 'sell', 'marker_position'] = data['High']*1.05
data
```

```python
# Filter the data that has no missing values
data[data['marker_position'].notnull()]
```

```python
# Plot graph
markers = ['^' if x == 'buy' else ('v' if x == 'sell' else None) for x in data['action']]
color_marker = ['green' if x == 'buy' else 'red' for x in data['action']]
plot_signal = mpf.make_addplot(data['marker_position'], type = 'scatter', markersize = 30, marker = markers, color = color_marker)

mpf.plot(data, type='candle',style = 'yahoo', addplot=[ema_short_line, ema_long_line, plot_signal], figsize=(16,9))
```

###### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; MACD: [https://th.tradingview.com/support/solutions/43000502344/](https://th.tradingview.com/support/solutions/43000502344/)

```python
# Create MACD line
macd = data['ema_short'] - data['ema_long']

# Create signal line
signal_line = macd.ewm(span=9, adjust = False, min_periods = 9).mean()
```
